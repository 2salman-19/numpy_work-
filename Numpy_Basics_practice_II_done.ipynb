# Q7

Statement: You are given a 3D NumPy array data of shape (5, 6, 7) representing the sensor readings from multiple sensors placed in different locations over a period of time. Each sensor provides measurements at different time points. Write a code snippet to calculate the following statistics for each sensor:

Total number of measurements

Average measurement value

Standard deviation of measurements

Minimum measurement value

Maximum measurement value

Sample Input:

```python
# Answer Here
import numpy as np

# Assuming data is the 3D array with shape (5, 6, 7)
# Create a sample array with random measurements for demonstration purposes
data = np.random.rand(5, 6, 7) * 100  # Random sensor readings between 0 and 100

# Initialize arrays to store the statistics for each sensor
total_measurements = data.shape[1] * data.shape[2]  # Total measurements for each sensor
average_measurements = np.mean(data, axis=(1, 2))  # Average measurement value
std_deviation = np.std(data, axis=(1, 2))           # Standard deviation of measurements
min_measurements = np.min(data, axis=(1, 2))        # Minimum measurement value
max_measurements = np.max(data, axis=(1, 2))        # Maximum measurement value

total_measurements
average_measurements
std_deviation
min_measurements
```
